```python
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
x = np.linspace(1, 10, 15)
```

```python
true_m = 0.5
noise = np.random.normal(size=len(x), loc=0, scale=0.3)
y_true = true_m * x + noise
y_pred = LinearRegression().fit(x.reshape(-1, 1), y_true).predict(x.reshape(-1, 1))
```

```python
plt.scatter(x.reshape(-1, 1), y_true)
plt.scatter(x.reshape(-1, 1), y_pred, color = "red")
```

```python
f = np.vectorize(lambda x, m: m * x)
```

```python
loss = [mean_squared_error(y_true, f(x, m)) for m in np.arange(0, 1, 0.01) ]
```

```python
plt.plot(np.arange(0, 1, 0.01), loss)
```

```python
from sklearn.metrics import mean_squared_error

x = np.linspace(1, 10, 15)
true_m = 0.5
noise = np.random.normal(size=len(x), loc=0, scale=0.3)
y_true = true_m * x + noise
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()

model.fit(x.reshape(-2, 1), y_true)

```

```python
x.reshape(-1, 1)
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x.reshape(-1, 1), y_true, test_size=1/3)
```

```python
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
mean_squared_error(y_pred, [1, *y_test])
```

```python
def mse(y_pred, y_test): 
    if len(y_pred) != len(y_test): 
        raise ValueError("len doesn't match")
    
    n = len(y_pred)
    
    square_error = 0 
    for y1, y2 in zip(y_pred, y_test): 
        square_error += (y1 - y2)**2
    
    return square_error / n
```

```python
mse(y_pred, [1, *y_test])
```
