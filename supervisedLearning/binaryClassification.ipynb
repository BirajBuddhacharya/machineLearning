```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, r2_score
from sklearn.preprocessing import OrdinalEncoder, StandardScaler
from sklearn.model_selection import train_test_split

df = pd.read_csv("placementdata.csv")
```

```python
df
```

```python
# data preprocessing
scaler = StandardScaler()
df[['CGPA', "Internships", "Projects", "Workshops/Certifications", "AptitudeTestScore", "SoftSkillsRating", 'SSC_Marks', 'HSC_Marks']] = scaler.fit_transform(df[['CGPA', "Internships", "Projects", "Workshops/Certifications", "AptitudeTestScore", "SoftSkillsRating", 'SSC_Marks', 'HSC_Marks']])

ordinalEncoder = OrdinalEncoder()
df[['PlacementStatus', 'ExtracurricularActivities', 'PlacementTraining']] = ordinalEncoder.fit_transform(df[['PlacementStatus', 'ExtracurricularActivities', 'PlacementTraining']])
```

```python
# model training
X, y = df.drop(['StudentID', 'PlacementStatus'], axis = 1), df['PlacementStatus']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=1/3)

model = LogisticRegression()

model.fit(X_train, y_train) # training model 
```

```python
y_pred = model.predict(X_test)
accuracy_score(y_pred, y_test)
```

```python
r2_score(y_pred, y_test)
```
