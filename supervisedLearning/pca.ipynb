```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv("../data/pca-analyze.csv")
```

```python
plt.figure(figsize=(10, 4))

plt.subplot(1, 2, 1)
plt.scatter(df.Height, df.Weight, c=df.index)
plt.xlabel('Height')
plt.ylabel('Weight')
plt.title('Height vs Weight')

plt.subplot(1, 2, 2)
encoder = StandardScaler()
df[['Height', 'Weight']] = encoder.fit_transform(df[['Height', 'Weight']])

pca = PCA(n_components=1)
x = pca.fit_transform(df[['Height', 'Weight']]).squeeze()
y = np.ones(len(x))

plt.scatter(x, y, c=df.index)
plt.xlabel('Principal Component 1')
plt.ylabel('Constant Value')
plt.title('PCA of Height and Weight')

plt.tight_layout()
plt.show()
```
