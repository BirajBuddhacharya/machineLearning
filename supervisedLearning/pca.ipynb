```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
from sklearn.decomposition import PCA
```

```python
df = pd.read_csv("../data/pca-analyze.csv")
df = df.sample(frac=1).reset_index(drop=True)
```

```python
df.plot(x = 'Height', y = 'Weight', kind = 'scatter')
```

```python
pca = PCA(n_components=1)
x = pca.fit_transform(df[['Height', 'Weight']]).squeeze()
y = np.ones(len(x))

plt.scatter(x, y)
```
