```python
model.fit(X_train, y_train)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
from sklearn.tree import DecisionTreeClassifier

# Generate 2D dataset
X, y = make_classification(n_samples=200, n_features=2, n_classes=2, n_redundant=0, random_state=42)

# Train classifier
clf = DecisionTreeClassifier(max_depth=3).fit(X, y)

# Plot decision boundary
xx, yy = np.meshgrid(np.linspace(X[:,0].min()-1, X[:,0].max()+1, 100),
                     np.linspace(X[:,1].min()-1, X[:,1].max()+1, 100))
Z = clf.predict(np.c_[xx.ravel(), yy.ravel()]).reshape(xx.shape)

plt.contourf(xx, yy, Z, alpha=0.3, cmap='coolwarm')  # Decision boundary
plt.scatter(X[:, 0], X[:, 1], c=y, edgecolors='k', cmap='coolwarm')  # Data points
plt.show()
```
